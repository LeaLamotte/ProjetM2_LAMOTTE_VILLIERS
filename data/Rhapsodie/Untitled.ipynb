# diagnostique Unités Illocutoires :

```python
rep_cour=os.getcwd()
ligne=re.compile("[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t([^\t]*)\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*\t([^\t]*)\t([^\t]*)\t[^\t]*\t[^\t]*\t[^\t]*\t[^\t]*")
copie={} #dico {fichier : liste des lignes}
for file in os.listdir(rep_cour): #pour tous les fichier tabulaire du repertoire courant
    if file.endswith(".tabular"):
        texte=open(file, encoding='UTF-8').read() #on l'ouvre et on le lit
        temp=ligne.findall(texte)#on en extrait les colonnes "UI", "Tmin" et "tmax"
        temp.remove(('IU', 'Tmin', 'Tmax'))
        while ('', '', '') in temp:
            temp.remove(('', '', ''))
        copie[file[5:-28]]=temp
```

```python
ok=0
for file in copie:
    b, l, u = 0, 0, 0
    for (ui, tmin, tmax) in copie[file]:
        if ui=='B':
            b=b+1
        elif ui=='L':
            l=l+1
        elif ui=='U':
            u=u+1
    if b>l:
        print(file, ':', b-l, 'B "de trop" par rapport au nombre de L.')
    elif b<l :
        print(file, ':', l-b, 'B "manquants" par rapport au nombre de L.')
    else : 
        ok=ok+1
print(ok)
    #print(file, ':', b, 'B,', l, 'L et', u, 'U.')
```
